```python
import pandas as pd
import numpy as np
orders = pd.read_csv('orders_raw.csv')
```

```python
# Displaying rows where city or/and email are empty.
orders[orders[['city', 'email']].isna().any(axis=1)]
```

```python
# Defining a regular expression pattern to validate email addresses and keeping only rows where email matches the pattern
import re
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
orders = orders[orders['email'].str.match(pattern, na=False, flags=re.IGNORECASE)]
orders
```

```python
# Changing Na values to 0 in quantity and to unknown in city
orders = orders.fillna({'quantity': 0, 'city':'unknown'})
orders[10:20]
```

```python
# Changing polish column names to english
orders = orders.rename(columns={"id_zamówienia": 'order_id', 'id_klienta': 'customer_id'})
orders
```

```python
# Showing duplicates in order_id column
orders[orders.duplicated(subset='order_id')]
```

```python
# Removing duplicates in order_id column
orders = orders.drop_duplicates(subset='order_id')
orders
```

```python
# Unifying vales in returned column (boolean values)
orders = orders.copy()
orders['returned'] = orders['returned'].replace({"yes": True, "YES": True, "1": True, 
                                                "no": False, "No": False, "0": False, 
                                                 "True": True, 'False': False})
orders[15:25]
```

```python
# Fixing typos
product_typos = {'m ouse': 'Mouse', 'Moniter': 'Monitor', 
                 'chargerr': 'Charger', 'Lapotp': 'Laptop', 
                 'Keybaord': 'Keyboard', 'Smar tphone': 'Smartphone', 
                 'headPhones': 'Headphones'}
orders['product'] = orders['product'].replace(product_typos)
orders[14:23]
```

```python
# Changing Na and negative values to median
orders.loc[orders['price']<0, 'price'] = pd.NA
orders["price"] = orders["price"].fillna(
    orders.groupby("product")["price"].transform("median")
)
orders[14:23]
```

```python
# Calculation of order vaue
orders["order_value"] = orders.price * orders.quantity
orders
```

```python
# Dividing the collection into designated baskets
cheap_names = ['very cheap', 'cheap', 'medium', 'expensive']
cheap_bins = [0, 50, 100, 1000, 1000000]
how_cheap = pd.cut(orders.price, cheap_bins, labels=cheap_names)
how_cheap
```

```python
# Dividing a set into quartiles
how_cheap_quartiles = pd.qcut(orders.price, 4, precision=2, labels=cheap_names)
how_cheap_quartiles
```

```python
# Adding new column how_cheap_quartiles and how_cheap
orders['how_cheap_quartiles'] = how_cheap_quartiles
orders['how_cheap'] = how_cheap
orders.head()
```

```python
# Fixing typos in city column
orders['city'] = orders['city'].str.strip().str.capitalize()
city_typos = {'Lodz': 'Łódź', 'Poznan': 'Poznań', 'Wroclav': 'Wrocław', 
              'Wroclaw': 'Wrocław', 'Krakow': 'Kraków', 'Gdansk': 'Gdańsk', 'War szawa': 'Warszawa'}
orders['city'] = orders['city'].replace(city_typos)
orders[14:23]
```

```python
# Designating a domain from an email and adding as a new column
orders['domain'] = orders['email'].str.extract(pattern, flags=re.IGNORECASE)[1]
orders.head()
```

```python
# Showing wich rows in coupon column constains 2024
orders['coupon'].str.contains('2024')
```

```python
# Changing type city and product columns to categorical
orders['city'] = pd.Categorical(orders['city'], ordered=True)
orders['product'] = pd.Categorical(orders['product'], ordered=True)
orders["city"].cat.categories
```

```python
# Showing dummies from city column
pd.get_dummies(orders["city"])
```

```python
# Changing on str and living only yyyy-mm-dd 
s = orders['order_date'].astype(str)
date_str = s.str.extract(r'(\d{4}-\d{2}-\d{2})', expand=False)

# Changigng type to date and live only adte without time
orders.loc[:, 'order_date'] = pd.to_datetime(date_str, errors='coerce').dt.date

orders.head()
```

```python
# Showing type and memory useage before change
print(repr(orders['quantity'].dtype))
print(repr(orders['returned'].dtype))
print(repr(orders['email'].dtype))
orders[['quantity', 'returned', 'email']].memory_usage(deep=True)
```

```python
# Changing to pandas types and showing type and memory useage  
orders['quantity'] = pd.Series(orders['quantity'], dtype="Int64")
orders['returned'] = pd.Series(orders['returned'], dtype="boolean")
orders['email'] = pd.Series(orders['email'], dtype="string")

print(repr(orders['quantity'].dtype))
print(repr(orders['returned'].dtype))
print(repr(orders['email'].dtype))
orders[['quantity', 'returned', 'email']].memory_usage(deep=True)
```
